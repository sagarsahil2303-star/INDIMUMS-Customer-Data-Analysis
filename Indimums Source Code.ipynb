```python
#Load Dataset
df = pd.read_excel("INDIMUMS_customer_orders.xlsx")

df.head()
```

```python
#Dataset Overview
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\nColumn Names:")
print(df.columns.tolist())

print("\nDataset Info:")
df.info()
```

```python
#Data Cleaning
print("Missing Values")
print(df.isnull().sum())

print("\nDuplicate Rows")
print(df.duplicated().sum())

print("\nDuplicate Order Numbers")
print(df["order_no"].duplicated().sum())
```

```python
#Unique Counts
print("Unique Customers:", df["customer_name"].nunique())
print("Unique Products:", df["product_purchased"].nunique())
print("Unique Cities:", df["city"].nunique())
```

```python
#Quantity Analysis
df["quantity_purchased"].describe()
print("Average Quantity:",
      round(df["quantity_purchased"].mean(),2))

print("Most Common Quantity:",
      df["quantity_purchased"].mode()[0])
```

```python
#Quantity Distribution
df["quantity_purchased"] \
    .value_counts() \
    .sort_index() \
    .plot(kind="bar")

plt.title("Quantity Purchased Distribution")
plt.xlabel("Quantity")
plt.ylabel("Orders")
plt.show()
```

```python
#City Analysis
top_cities = df["city"].value_counts()

top_cities
top_cities.head(5).plot(kind="bar")

plt.title("Top Cities by Orders")
plt.xlabel("City")
plt.ylabel("Orders")
plt.show()
```

```python
#Product Analysis
top_products = df["product_purchased"].value_counts()

top_products
top_products.plot(kind="bar")

plt.title("Top Products")
plt.xlabel("Product")
plt.ylabel("Orders")
plt.show()
```

```python
#Marketing Channel Analysis
top_channels = df["marketing_channel"].value_counts()

top_channels
top_channels.plot(kind="bar")

plt.title("Marketing Channel Performance")
plt.xlabel("Channel")
plt.ylabel("Orders")
plt.show()
```
